```python
import pandas as pd
import os
```

```python
# Create an empty dataframe
df = pd.DataFrame(columns=['image_path', 'label'])
```

```python
# Define the labels/classes
labels = {"E:\LiveProject\Deep_Learning\mobile phone Damage\good" : 'Good',
          "E:\LiveProject\Deep_Learning\mobile phone Damage\ground_truth_1" : 'ground_truth_1',
          "E:\LiveProject\Deep_Learning\mobile phone Damage\ground_truth_2" : 'ground_truth_2',
          "E:\LiveProject\Deep_Learning\mobile phone Damage\oil" : 'Oil',
          "E:\LiveProject\Deep_Learning\mobile phone Damage\scratch" : 'Scratch',
          "E:\LiveProject\Deep_Learning\mobile phone Damage\stain" : 'Stain',
           }
```

```python
# Loop over the train, test, and val folders and extract the image path and label
for folder in labels:
    for image_name in os.listdir(folder):
        image_path = os.path.join(folder, image_name)
        label = labels[folder]
        df = df.append({'image_path': image_path, 'label': label}, ignore_index=True)
```

```python
# Save the data to a CSV file
df.to_csv('mobile_phone_damage.csv', index=False)
```

```python
from PIL import Image
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pandas as pd
import os
```

```python
# Split the dataset into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.25, random_state=42)   
```

```python
#Define the image data generators for training and testing
train_datagen = ImageDataGenerator(rescale=1./255, 
                                   shear_range=0.2, 
                                   zoom_range=0.2, 
                                   horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Generate augmented images for training
train_generator = train_datagen.flow_from_dataframe(dataframe=train_df,
                                                    x_col="image_path",
                                                    y_col="label",
                                                    target_size=(100, 100),
                                                    batch_size=32,
                                                    class_mode="categorical")
```

```python
# Generate augmented images for testing
test_generator = test_datagen.flow_from_dataframe(dataframe=test_df,
                                                  x_col="image_path",
                                                  y_col="label",
                                                  target_size=(100, 100),
                                                  batch_size=32,
                                                  class_mode="categorical")
```

```python
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import MobileNetV3Large
```

```python
# Create the MobileNetV3Larget base model
base_model = MobileNetV3Large(include_top=False, input_shape=(224, 224, 3))
```

```python
# Build the model architecture
model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(6, activation='softmax'))
```

```python
# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history_v2 = model.fit(train_generator, epochs=25, validation_data=test_generator)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_generator)
```

```python
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# save it as a h5 file
from tensorflow.keras.models import load_model
model.save('mobile_damage.h1')
```
